```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Función para recrear derivación genética en una población, utilizando el método de Montecarlo 
# params: población, tasa de mutación
# return: nueva población

def montecarlo_deriva_genetica(poblacion, tasa_mutacionA, tasa_mutacionB):
    nueva_poblacion = []

    for alelo in poblacion:
        mutarA = random.uniform(0, 1) < tasa_mutacionA
        mutarB = random.uniform(0, 1) < tasa_mutacionB
        if alelo == 'A' and mutarA:
            nueva_poblacion.append('B')
        elif alelo == 'B' and mutarB:
            nueva_poblacion.append('A')
        else:
            nueva_poblacion.append(alelo)

    return nueva_poblacion
```

```python
# Función para la simulación en base al número de generaciones
# params: población inicial, generaciones, tasa de mutación
# return: frecuencias, las frecuencias de ambos alelos 'A' y 'B'

def montecarlo_simulacion_genetica(poblacion_inicial, generaciones, tasa_mutacionA, tasa_mutacionB):
    poblacion_actual = poblacion_inicial.copy()
    frecuencias = {'A': [], 'B': []}

    for generacion in range(generaciones):
        poblacion_actual = montecarlo_deriva_genetica(poblacion_actual, tasa_mutacionA, tasa_mutacionB)

        # Calcular frecuencias y agregarlas al diccionario
        frecuencia_A, frecuencia_B = analizar_genetica(poblacion_actual)
        frecuencias['A'].append(frecuencia_A)
        frecuencias['B'].append(frecuencia_B)
        
    return frecuencias
```

```python
# Función de análisis de la genética
# params: población
# return: frecuencia_A, frecuencia_B; frecuencias individuales de alelo 'A' y 'B'

def analizar_genetica(poblacion):
    frecuencia_A = poblacion.count('A') / len(poblacion)
    frecuencia_B = poblacion.count('B') / len(poblacion)
    
    return frecuencia_A, frecuencia_B
```

```python
# Función para generar el gráfico de frecuencia de alelos
# params: frecuencias, generaciones
def graficar_evolution(frecuencias, generaciones, title):
    plt.subplot(2, 1, 1)
    plt.plot(generaciones, frecuencias['A'], label='Alelo A')
    plt.plot(generaciones, frecuencias['B'], label='Alelo B')
    plt.xlabel('Generación')
    plt.ylabel('Frecuencia del Alelo')
    plt.legend()
    plt.title(title)

    # Establecer límites y marcas de intervalo en el eje y
    plt.ylim(-0.05, 1.05)
    plt.yticks(np.arange(0, 1.1, 0.2))

    plt.tight_layout()
    plt.show()
```

```python
# Parámetros
generaciones = 200
tasa_mutacion_B = 0.01
tasa_mutacion_A0 = 0.01
tasa_mutacion_A1 = 0.05
tasa_mutacion_A2 = 0.10
tasa_mutacion_A3 = 0.15
tasa_mutacion_A4 = 0.20
tasa_mutacion_A5 = 0.25

# Población inicial
poblacion_inicial = ['A', 'B', 'A', 'A', 'B', 'B', 'A', 'B', 'A', 'A']

# Ejecutar simulación con Montecarlo usando random.uniform
frecuencias_0 = montecarlo_simulacion_genetica(poblacion_inicial, generaciones, tasa_mutacion_A0, tasa_mutacion_B)
frecuencias_1 = montecarlo_simulacion_genetica(poblacion_inicial, generaciones, tasa_mutacion_A1, tasa_mutacion_B)
frecuencias_2 = montecarlo_simulacion_genetica(poblacion_inicial, generaciones, tasa_mutacion_A2, tasa_mutacion_B)
frecuencias_3 = montecarlo_simulacion_genetica(poblacion_inicial, generaciones, tasa_mutacion_A3, tasa_mutacion_B)
frecuencias_4 = montecarlo_simulacion_genetica(poblacion_inicial, generaciones, tasa_mutacion_A4, tasa_mutacion_B)
frecuencias_5 = montecarlo_simulacion_genetica(poblacion_inicial, generaciones, tasa_mutacion_A5, tasa_mutacion_B)

# Graficar las frecuencias a lo largo de las generaciones
generaciones = list(range(generaciones))
graficar_evolution(frecuencias_0, generaciones, 'Evolución de la Frecuencia de Alelos con mutacion 0.01')
graficar_evolution(frecuencias_1, generaciones, 'Evolución de la Frecuencia de Alelos con mutacion 0.05')
graficar_evolution(frecuencias_2, generaciones, 'Evolución de la Frecuencia de Alelos con mutacion 0.10')
graficar_evolution(frecuencias_3, generaciones, 'Evolución de la Frecuencia de Alelos con mutacion 0.15')
graficar_evolution(frecuencias_4, generaciones, 'Evolución de la Frecuencia de Alelos con mutacion 0.20')
graficar_evolution(frecuencias_5, generaciones, 'Evolución de la Frecuencia de Alelos con mutacion 0.25')
```
